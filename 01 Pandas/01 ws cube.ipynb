### Merge, Join and Concatenate in Pandas

```python
data1 = {
    "Emp Id": [
        "E01", "E02", "E03", "E04", "E05", "E06"
    ],
    "Name": [
        "Awais", "Ali", "Saad", "Imran", "Haroon", "Abbas"
    ],
    "Age": [23,31,35,64,20,43]
}

data2 = {
    "Emp Id": ["E01", "E02", "E03", "E04", "E05", "E06"],
    "Salary": [35200, 40000, 5000, 20000, 53000, 2200]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
print(df1)
print(df2)
```

```python
#Merge
print(pd.merge(df1,df2, on = "Emp Id", how = "left"))
```

```python
print(pd.merge(df1,df2, on = "Emp Id", how = "right"))
```

```python
print(pd.join([df1,df2]))
```

```python
print(pd.concat([df1,df2]))
```
